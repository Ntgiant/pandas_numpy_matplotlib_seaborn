
### Grouping

As you know, grouping data in Pandas works as follows:


```python
columns_to_show = ["Total day minutes", "Total eve minutes", "Total night minutes"]

df.groupby(["Churn"])[columns_to_show].describe() 
```

Let’s do the same thing, but slightly differently by passing a list of functions to `agg()`:

```python
columns_to_show = ["Total day minutes", "Total eve minutes", "Total night minutes"]

df.groupby(["Churn"])[columns_to_show].agg([np.mean, np.std, np.min, np.max])
```

```python
df.groupby(["Churn"])[columns_to_show].agg(["mean", "std","min", "max"])
```


### DataFrame transformations

Calculate the total number of calls for all users. Create the `total_calls` Series. (total calls is total day/eve/night/intl calls combined)

```python
cols = ['Total day minutes',
       'Total day calls', 'Total day charge', 'Total eve minutes',
       'Total eve calls', 'Total eve charge', 'Total night minutes',
       'Total night calls', 'Total night charge', 'Total intl minutes',
       'Total intl calls', 'Total intl charge']
```

```python
df['total_calls'] = df[cols].apply(np.sum, axis=1)
df.head()
```

Delete the column that you just created.  (sepakan aracy jurn enq qcum ....)

```python
df.drop('total_calls', axis=1, inplace=True)
```
