**Find genres in the columns**

* Give me 10 most common genres for movies and 10 most common genres. (Pay attention that some rows contain multiple values)


```python
genres = {}

for i, val in enumerate(df['listed_in']):
    val = df['listed_in'].apply(lambda x: x.split(', '))
    for j in val[i]:
        if j in genres.keys():
            genres[j] += 1
        else:
            genres[j] = 1
    
```

```python
### Way 1

sort_gen = sorted(genres.values())[-10:]
sort_gen = sort_gen[::-1]


for i in sort_gen:
    for j in genres:
        if genres[j] == i:
            print(f"{j} -> {i}")
            print('*')
```

**Show distribution of year column for each type?**

**What is median value of year for each type? Or are there outliers?**


```python
med = df[df['type'] == 'Movie'].release_year.median()

plt.figure(figsize=(12,10))
sns.distplot(x= df[df['type'] == 'Movie'].release_year)
plt.axvline(x= df[df['type'] == 'Movie'].release_year.median(), c='red', ls='--', lw=1.2, label=f'median {med}')
#plt.axvline(x= df[df['type'] == 'Movie'].release_year.mean(), c='y', ls='--', lw=1.2, label='mean')

plt.xticks([1940, 1960, 1980, 2000, 2005, 2010, 2016, 2020], rotation=45)
plt.legend(loc='upper left', fontsize=15)
plt.title('Year distribution for Movies');
```

```python
med2 = df[df['type'] == 'TV Show'].release_year.median()
#fig, ax = 
plt.subplots(figsize=(12,10))
sns.distplot(x= df[df['type'] == 'TV Show'].release_year)
plt.axvline(x= df[df['type'] == 'TV Show'].release_year.median(), c='red', ls='--', label=f'median {med2}')
#plt.axvline(x= df[df['type'] == 'TV Show'].release_year.mean(), c='y', ls=':', label='mean')

#ax.set_xticks([1940, 1960, 1980, 2000, 2005, 2010, 2015, 2020])
plt.xticks(np.linspace(1900, 2030, 11), rotation=45)
plt.legend(loc='upper left', fontsize=15)
plt.title('Year distribution for TV Shows');
```

```python
#### the median release year for movies is 2016 and for TV shows is 2018. They both had boom
###  of number of releases after the year 2000. For mivoes this release boom is more gradual, 
###  which occured gradually from the year 2000-2010, while for TV shows the boom is more abrupt,
###  which ocured around the years of 2004-2006.
```
